```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# 1. Create the DataFrame
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

# --- Task 1: Clean the Data ---

# Impute age with median
df['age'] = df['age'].fillna(df['age'].median())

# Impute city with mode (mode returns a Series, so we take the first index)
df['city'] = df['city'].fillna(df['city'].mode()[0])

# Remove duplicate rows
df = df.drop_duplicates()

# --- Task 2: Encode Categorical Columns ---

# Label Encoding for gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

# One-Hot Encoding for city
df = pd.get_dummies(df, columns=['city'], prefix='city')

print("--- Cleaned and Encoded DataFrame ---")
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Prepare numerical columns for scaling
numerical_cols = ['age', 'annual_income']

# Min-Max Scaling
min_max_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numerical_cols] = min_max_scaler.fit_transform(df[numerical_cols])

print("\n--- Results of Min-Max Scaling ---")
print(df_minmax[numerical_cols])

# Standardization (Z-score normalization)
std_scaler = StandardScaler()
df_std = df.copy()
df_std[numerical_cols] = std_scaler.fit_transform(df[numerical_cols])

print("\n--- Results of Standardization ---")
print(df_std[numerical_cols])
,
```
